```python
# Apriori

support_threshold = 0.3

supports = {crit: len(passing_items)/len(items) for crit, passing_items in passing_items_for_crits.items()}
# Prune all critiques that don't meet the support threshold
supports_exceeding_thres = {crit: support for crit, support in supports.items() if support > support_threshold}

for crit, support in supports_exceeding_thres.items():
    print("{}: {}".format(crit, support))
```

```python
from framework.critique import CompoundPropCritique
# First pass of apriori
# Combine remaining critiques in compound critiques, check their support and prune by threshold
current_pass = []
for i, crit_a in enumerate(supports_exceeding_thres):
    for crit_b in list(supports_exceeding_thres.keys())[:i]:
        # TODO later beter uitzoeken maar denk dat best wordt uitgesloten dat 2 critiques op zelfde property niet mogen worden gecombineerd
        if crit_a.prop_name != crit_b.prop_name:
            current_pass.append(CompoundPropCritique([crit_a, crit_b]))
current_pass_support = get_passing_items_for_critiques(current_pass, items)
current_pass_support = {crit: len(passing_items)/len(items) for crit, passing_items in current_pass_support.items() if len(passing_items)/len(items) > support_threshold}

for crit, support in current_pass_support.items():
    print("{}: {}".format(crit, support))
    
current_pass = [crit for crit in current_pass_support]
```

```python
def get_common_crits_of_compounds(com_crit_a, com_crit_b):
    return set(com_crit_a.prop_critiques).intersection(com_crit_b.prop_critiques)

def get_uncommon_crits_of_compounds(com_crit_a, com_crit_b):
    return set(com_crit_a.prop_critiques).symmetric_difference(com_crit_b.prop_critiques)

def combine_compound_crits(com_crit_a, com_crit_b):
    return CompoundPropCritique(set(com_crit_a.prop_critiques).union(com_crit_b.prop_critiques))

def have_overlapping_props_for_crits(crits):
    return len(set(crit.prop_name for crit in crits)) < len(crits)
# Then the process is repeated for each critique of size k in the last pass that has k-1 common critiques with another item until no larger compound can be created (none pass threshold, or no left to combine)
# Find pairs with k-1 common critiques
k = 2
last_pass = current_pass
current_pass = list()
for i, crit_a in enumerate(last_pass):
    for crit_b in last_pass[:i]:
        common_crits = get_common_crits_of_compounds(crit_a, crit_b)
        # Check if has common k-1 common critiques
        if len(common_crits) == k - 1:
            # Make sure that none of the uncommon critiques cover the same property (see earlier TODO)
            uncommon_crits = get_uncommon_crits_of_compounds(crit_a, crit_b)
            if not have_overlapping_props_for_crits(uncommon_crits):
                # Combine critiques in one
                new_comp_crit = combine_compound_crits(crit_a, crit_b)
                current_pass.append(new_comp_crit)
            else:
                print(crit_a, crit_b)
current_pass_support = get_passing_items_for_critiques(current_pass, items)
print(len(current_pass_support))
current_pass_support = {crit: len(passing_items)/len(items) for crit, passing_items in current_pass_support.items() if len(passing_items)/len(items) > support_threshold}

# for crit, support in current_pass_support.items():
#     print("{}: {}".format(crit, support))
# current_pass = [crit for crit in current_pass_support]
```

```python
# Kan waarschijnlijk geen kwaad om eens wat testen voor te schrijven
def apriori(anchor_crits, items, support_threshold, confidence_threshold = 0, max_compound_length=None):
    # Convert to compound critiques (for hashing recoverability in later passes and to allow combination)
    anchor_crits_as_comp = [CompoundPropCritique([crit]) for crit in anchor_crits]
    current_pass_support = {crit: len(passing_items)/len(items) for crit, passing_items in get_passing_items_for_critiques(anchor_crits_as_comp, items).items()}    

    # Filter by support only  
    current_pass = set(crit for crit, support in current_pass_support.items() if support >= support_threshold)
    
    # Start iterative loop
    current_pass_comp_length = 1
    dynamic_comp_crits = list()
    last_pass = list(current_pass)
    last_pass_support = current_pass_support
    max_length_reached = True if max_compound_length == 1 else False
    # While last_pass created at least two critiques so that one larger one could be created
    while(len(last_pass) >= 2 and not max_length_reached):
        current_pass_comp_length += 1
        current_pass = set()
        current_pass_support = dict()
        
        # Go over all combinations of critiques in the last pass (order of crits doesnt matter)      
        for i, crit_a in enumerate(last_pass):
            for crit_b in last_pass[:i]:
                # Check if the required amount of common critiques are present in the compound critiques, which is length of critiques in previous previous pass - 1 
                if len(get_common_crits_of_compounds(crit_a, crit_b)) == current_pass_comp_length - 2:
                    # Make sure that to be combined critiques (not common) don't cover the same property TODO voelt correct maar niet zeker van
                    uncommon_crits = get_uncommon_crits_of_compounds(crit_a, crit_b)
                    if not have_overlapping_props_for_crits(uncommon_crits):
                        # Create new compound critique of size k+1 from 2 critiques of size k with k - 1 common prop critiques                       
                        new_comp_crit = combine_compound_crits(crit_a, crit_b)
                        # Check if a permutation of this critique has already been added                        
                        if new_comp_crit not in current_pass:
                            # Calculate support new compound critique
                            support = len(get_passing_items_for_critique(new_comp_crit, items)) / len(items)
                            current_pass_support[new_comp_crit] = support
                            # Calculate confidence with support and supports of parts
                            confidence = support / (last_pass_support[crit_a] * last_pass_support[crit_b])
                            
                            # Filter based on support and confidence
                            if support >= support_threshold:
                                current_pass.add(new_comp_crit)


        # Add all satisfactory new compound critiques of the pass to the total list of all passes
        dynamic_comp_crits = dynamic_comp_crits + list(current_pass)
        # Make data of current pass available for next pass        
        last_pass = list(current_pass)
        last_pass_support = current_pass_support
        
        # Check if max compound length has been set, and if so, if it has been reached
        if max_compound_length and current_pass_comp_length == max_compound_length:
            max_length_reached = True
    return dynamic_comp_crits
```

```python
from framework.critique.unit import NotCritique

def blacklist(crits, blacklist):
    return [crit for crit in crits if not(crit.prop_name in blacklist and type(crit.critique) in blacklist[crit.prop_name])]

ignore_numerical_Not_crits = blacklist(prop_unit_critiques, {"LotFrontage": [NotCritique], "LotArea": [NotCritique], "Alley": [NotCritique]})
dyn = apriori(ignore_numerical_Not_crits, items, 0.1)
```
